```python
# We choose to have 25 sites
N = 25

T = 0.0
kB = 1

t = 1
delta = t

mu = 0 * t

H = get_kiatev_hamiltonian(N, t=t, mu=mu)
bdg_H = get_bdg_hamiltonian(H, delta=np.zeros(np.shape(H)))

for i in range(1):
    last_delta = delta
    energy_array, eigen_vectors = np.linalg.eig(bdg_H)
    delta = update_delta(N, delta, energy_array, eigen_vectors, T=T)
    # print("Diff: ", np.abs(delta - last_delta))
    bdg_H = get_bdg_hamiltonian(H, delta=np.zeros(np.shape(H)))


for i, e in enumerate(np.real(energy_array)):
    plt.scatter(i%25,e)

plt.show()
```
