```python
import numpy as np

import matplotlib.pyplot as plt
```

```python
def autocorr(x):
    result = np.correlate(x, x, mode='full')
    return result[np.size(result)//2:]


x_range = np.linspace(0, 1, 150)

f = 20
data = np.cos(f * (2 * np.pi) * x_range )

ac = autocorr(data) / 100

plt.plot(x_range, ac, label="autocorrelation")
plt.plot(x_range, data, label="data")

plt.legend()
```
